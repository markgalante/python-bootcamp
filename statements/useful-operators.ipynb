# Operators

### `range(start, end, jump)`
```python
list(range(8)) # [0, 1, 2, 3, 4, 5, 6, 7]
list(range(4,8)) # [4, 5, 6, 7]
list(range(4,8,2)) # [4,6]


# Basic
for num in range(8):
  print(num)
0...
...7

# Start and end
for num in range(3, 10):
  print(num)
3...
...9

# Start, end and range
for num in range(2, 11, 2):
  print(num)
2
4
6
8
10
```

### `enumerate()`
Returns a tuple of (index, item)
```python
word = 'abcdef'
for item in enumerate(word):
  print(item)
(0, 'a')
(1, 'b')
(2, 'c')
(3, 'd')
(4, 'e')
(5, 'f')

word = ["Mark", "Erez", "Martin", "Shai", "Xola"]
for idx, person in enumerate(word):
  print(idx)
0...4

word = ["Mark", "Erez", "Martin", "Shai", "Xola"]
for idx, person in enumerate(word):
  print(person)
Mark...
...Xola
```

### `zip()`
**NOTE:** only zips up until the shortest list.

```python
list1 = [1,2,3]
list2 = ['a', 'b', 'c']

# you can also do "tuple unpacking"
for item in zip(list1, list2):
  print(item)
#Prints:
(1, 'a')
(2, 'b')
(3, 'c')
```

### `in`
Checks if object is in a list
```python
'x' in 'xyz' # True
'x' in 'abc' # False
1 in [1,2,3] # True
4 in [1,2,3] # False

d = {'key1': 100}
'key1' in d # True
'key1' in d.keys() # True
'key2' in d # False
100 in d.values() # True

```

### `min()`
***ONLY*** for numbers
```python
nums = [500, 400, 300, 200, 100, 900, 800, 700, 600]
min(nums) # 100
```


### `max()`
```python
nums = [500, 400, 300, 200, 100, 900, 800, 700, 600]
max(nums) # 900
```


### `random` library
**Shuffling a list**
```python
from random import shuffle
nums = [1,2,3,4,5,6,7,8,9]
shuffle(nums)
nums # [6, 1, 5, 9, 4, 2, 3, 8, 7]
```
**Grabbing a random integer from a range of numbers**
```python
from random import randint
randint(0, 100)
```
